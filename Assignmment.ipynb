```python
import pandas as pd
df=pd.read_csv("Supermart_Grocery_Sales.csv.xls")
df.head(10)
```

```python
result=df[df["Profit"]>150][["Customer Name","Category","Sales"]]
print(result)
```

```python
df["Sales"].fillna(df["Sales"].mean(),inplace=True)
df["Discount"].fillna(method="bfill",inplace=True)
df["Profit"].fillna(method="bfill",inplace=True)
print(df.isnull().sum())
```

```python
df["Sales_St_Norm"]=(df["Sales"]-df["Sales"].min()/df["Sales"].max()-df["Sales"].min())
print(df[["Sales","Sales_St_Norm"]].head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
transactions=df.groupby("Order ID")["Sub Category"].apply(list).values
te=TransactionEncoder()
te_array=te.fit(transactions).transform(transactions)
transactions_df=pd.DataFrame(te_array,columns=te.columns_)
frequent_itemsets=apiori(transaction_df,min_support=0.02,use_colnames=True)
print("frequent Itemsets")
print(frequent_itemsets)
rules=association_rules(frequent_itemsets,metric="confidence",min_thershold=0.5)
print("Assocation Rules")
print(rules[['antecedents','consequents','support','confidence']])
plt.figure()
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel("support")
plt.ylabel("confidence")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df["Sales"]=df["Sales"].fillna(df["Sales"].mean())
df["Discount"]=df["Discount"].bfill()
df["Profit"]=df["Profit"].ffill()
df=df.dropna()
le=LabelEncoder()
for col in df.select_dtypes(include=["object","string"]).columns:
    df[col]=le.fit_transform(df[col])
x=df.drop("Region",axis=1)
y=df["Region"]
print("total :",x.isnull().sum().sum())
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print("Accuracy",accuracy_score(y_test,y_pred))
```
